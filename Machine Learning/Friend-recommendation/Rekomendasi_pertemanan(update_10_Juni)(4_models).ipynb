**Saving Model using TF (coba-coba)**

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Concatenate, Dense

# Read the data
df = pd.read_csv("/content/Data generated.csv")
data = pd.DataFrame({'userid': df['userid'], 'instansi': df['instansi'], 'pekerjaan': df['pekerjaan'],
                     'domisili': df['domisili'], 'hobi': df['hobi']})

# Create user and item indices
user_indices = data['userid'].astype('category').cat.codes.values
instansi_indices = data['instansi'].astype('category').cat.codes.values
pekerjaan_indices = data['pekerjaan'].astype('category').cat.codes.values
domisili_indices = data['domisili'].astype('category').cat.codes.values
hobi_indices = data['hobi'].astype('category').cat.codes.values
item_indices = np.arange(len(data))

# Define the model architecture
num_users = len(data['userid'].unique())
num_instansi = len(data['instansi'].unique())
num_pekerjaan = len(data['pekerjaan'].unique())
num_domisili = len(data['domisili'].unique())
num_hobi = len(data['hobi'].unique())
embedding_dim = 32

user_input = Input(shape=(1,))
instansi_input = Input(shape=(1,))
pekerjaan_input = Input(shape=(1,))
domisili_input = Input(shape=(1,))
hobi_input = Input(shape=(1,))

user_embedding = Embedding(num_users, embedding_dim, embeddings_initializer='he_normal')(user_input)
instansi_embedding = Embedding(num_instansi, embedding_dim, embeddings_initializer='he_normal')(instansi_input)
pekerjaan_embedding = Embedding(num_pekerjaan, embedding_dim, embeddings_initializer='he_normal')(pekerjaan_input)
domisili_embedding = Embedding(num_domisili, embedding_dim, embeddings_initializer='he_normal')(domisili_input)
hobi_embedding = Embedding(num_hobi, embedding_dim, embeddings_initializer='he_normal')(hobi_input)

user_vec = tf.keras.layers.Flatten()(user_embedding)
instansi_vec = tf.keras.layers.Flatten()(instansi_embedding)
pekerjaan_vec = tf.keras.layers.Flatten()(pekerjaan_embedding)
domisili_vec = tf.keras.layers.Flatten()(domisili_embedding)
hobi_vec = tf.keras.layers.Flatten()(hobi_embedding)

user_preferences = Concatenate()([instansi_vec, pekerjaan_vec, domisili_vec, hobi_vec])
rating_pred = Dense(1)(user_preferences)

model = Model(inputs=[user_input, instansi_input, pekerjaan_input, domisili_input, hobi_input], outputs=rating_pred)

# Compile the model
model.compile(loss='mse', optimizer='adam')

# Split the data into train and test sets
train_ratio = 0.8
train_size = int(train_ratio * len(data))

train_user_input = user_indices[:train_size]
train_instansi_input = instansi_indices[:train_size]
train_pekerjaan_input = pekerjaan_indices[:train_size]
train_domisili_input = domisili_indices[:train_size]
train_hobi_input = hobi_indices[:train_size]
train_ratings = data['userid'][:train_size].values

test_user_input = user_indices[train_size:]
test_instansi_input = instansi_indices[train_size:]
test_pekerjaan_input = pekerjaan_indices[train_size:]
test_domisili_input = domisili_indices[train_size:]
test_hobi_input = hobi_indices[train_size:]
test_ratings = data['userid'][train_size:].values

# Train the model
model.fit([train_user_input, train_instansi_input, train_pekerjaan_input, train_domisili_input, train_hobi_input], train_ratings,
          validation_data=([test_user_input, test_instansi_input, test_pekerjaan_input, test_domisili_input, test_hobi_input], test_ratings),
          epochs=10, batch_size=32)

# Convert the model to TensorFlow Lite
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()
```

# Friends Recommendation Model (2) TF

```python
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("/content/Data generated.csv")

data = pd.DataFrame({'userid': df['userid'], 'instansi': df['instansi'], 'pekerjaan': df['pekerjaan'], 'domisili': df['domisili'], 'hobi': df['hobi']})

def create_model(input_size):
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(32, activation='relu', input_shape=(input_size,)),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])
    return model

def prepare_data(user_id):
    user_preferences = {
        'instansi': data[data['userid'] == user_id]['instansi'].values[0],
        'pekerjaan': data[data['userid'] == user_id]['pekerjaan'].values[0],
        'domisili': data[data['userid'] == user_id]['domisili'].values[0],
        'hobi': data[data['userid'] == user_id]['hobi'].values[0]
    }

    # Encode categorical variables
    encoder = LabelEncoder()
    encoded_data = data.copy()
    encoded_data['instansi'] = encoder.fit_transform(encoded_data['instansi'])
    encoded_data['pekerjaan'] = encoder.fit_transform(encoded_data['pekerjaan'])
    encoded_data['domisili'] = encoder.fit_transform(encoded_data['domisili'])
    encoded_data['hobi'] = encoder.fit_transform(encoded_data['hobi'])

    input_data = []
    output_data = []

    for i in range(len(encoded_data)):
        if encoded_data['userid'][i] != user_id:
            similar_user = (
                int(encoded_data['instansi'][i] == user_preferences['instansi']) or
                int(encoded_data['pekerjaan'][i] == user_preferences['pekerjaan']) or
                int(encoded_data['domisili'][i] == user_preferences['domisili']) or
                int(encoded_data['hobi'][i] == user_preferences['hobi'])
            )
            input_data.append([encoded_data['instansi'][i], encoded_data['pekerjaan'][i], encoded_data['domisili'][i], encoded_data['hobi'][i]])
            output_data.append(similar_user)

    return tf.constant(input_data), tf.constant(output_data)

# Prepare the data
user_id = 5492
x_train, y_train = prepare_data(user_id)

# Create and train the model
input_size = 4  # Number of input features
model = create_model(input_size)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=10)

# Convert the model to TensorFlow Lite format
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

# Save the TensorFlow Lite model to a file
from google.colab import drive
drive.mount('/content/drive')
with open('/content/drive/MyDrive/Capstone/saved model/model2.tflite', 'wb') as f:
    f.write(tflite_model)
```
